# Koenig2012 - Glucose model

Reproduction of part of the results from the original publication using `tellurium` and `libroadrunner`.  
The publication is available as [PDF](Koenig2012.pdf) and via http://www.ncbi.nlm.nih.gov/pubmed/22761565.

```python
from __future__ import print_function, division
%matplotlib inline
from IPython.display import Image
from IPython.core.display import HTML 

model_file = 'Hepatic_glucose_3.xml'
```

```python
# load model
import tellurium as te
r = te.loads(model_file)
```

```python
r.reset()
r.simulate(start=0, end=100, steps=100)
r.plot()


```

```python
# simulate
r.resetAll()
r.reset()
r.resetToOrigin()
# r.timeCourseSelections = 
r.timeCourseSelections = ['time'] + ['[{}]'.format(sid) for sid in r.getFloatingSpeciesIds()]
res = r.simulate(start=0, end=10, steps=100)
```

```python
# create the figure
from matplotlib import pylab as plt
fig = plt.figure(num=None, figsize=(20, 18), dpi=80, facecolor='w', edgecolor='k')
# fig, axis = plt.subplots(nrows=5, ncols=6, figsize=(20, 18), dpi=200, facecolor='w', edgecolor='k')
# fig.tight_layout()

# set the spaces between the subplots
fig.subplots_adjust(hspace=.3)
fig.subplots_adjust(wspace=.5)

for k, sid in enumerate(sorted(r.timeCourseSelections[1:])):
    print(k, sid)
    plt.subplot(5,6,k+1)
    plt.plot(res['time'], res[sid], linewidth=2, color='black')
    plt.title('{}'.format(sid), fontweight='bold')
    plt.ylabel('{} [mM]'.format(sid), fontweight='bold')
    plt.xlabel('time [h]'.format(sid), fontweight='bold')
    # plt.xlim([0, max_atp*1.05])
    # plt.ylim([0, max_adp*1.05])
    
plt.show()
fig.savefig('Hepatic_glucose_timecourse.png', dpi=fig.dpi)
```

```python
# calculate dose response curve of the model
import numpy as np
glc_vec = np.linspace(2, 20)
r.timeCourseSelections = ["ins", "glu", "epi", "gamma"]
r.simulate(0,1,steps=1)

for glc in glc_vec:
    r['[glc]'] = glc
    print(glc, r['[glc]'], r['ins'])
```
